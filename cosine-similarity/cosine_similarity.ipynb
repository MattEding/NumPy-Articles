```python
import functools

import matplotlib.pyplot as plt
import numpy as np
```

```python
size = 0.3
overhang = 0.2
width = 0.05

axisline = functools.partial(plt.arrow, width=0.05, head_width=size/2, head_length=size/2, color='black')
point = functools.partial(plt.plot, marker='o', markersize='15', markeredgecolor='black')
vector = functools.partial(plt.arrow, overhang=overhang, width=width, head_width=size, head_length=size)
alphavector = functools.partial(vector, alpha=0.5)

np.random.seed(19)
points = np.random.random_sample(12).reshape(-1, 2) * 8

centroid = points.mean(0)
thetas = [np.arctan2(y, x) for (x, y) in points]
thetas

plt.figure(figsize=(10, 10))
axis = plt.subplot(111, aspect='equal')


for i, pt in enumerate(sorted(points, key=lambda p: np.arctan2(p[1], p[0]))):
    x, y = pt

    if i == 1:
        point(x, y, color='blue')
        alphavector(0, 0, x * 10, y * 10, color='blue')
    else:
        point(x, y, color='black')

x, y = centroid
point(x, y, color='blue')
vector(0, 0, x * 10, y * 10)


axisline(0, 0, 0, 8)
axisline(0, 0, 8, 0)

plt.xlim(-0.09, 8.2)
plt.ylim(-0.09, 8.2)
plt.axis('off')
plt.show()
```
